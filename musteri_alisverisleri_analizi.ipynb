```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("musteri_alisverisleri_data.csv" , encoding = "latin-1")
```

```python
df
```

```python
df.info()
```

```python
df.columns = ["fatura_no","musteri_id","cinsiyet","yas","kategori","miktar","fiyat","odeme_turu","fatura_tarih","avm"]
```

```python
df["cinsiyet"].unique()
```

```python
df["kategori"].unique()
```

```python
df["odeme_turu"].unique()
```

```python
df["avm"].unique()
```

```python
df
```

```python
df
```

```python
df["cinsiyet"] = df["cinsiyet"].map({"Male":"Erkek" , "Female" : "Kadın"})
```

```python
df["kategori"] = df["kategori"].map({"Clothing":"Giyim" , "Shoes":"Ayakkabı" , "Books" : "Ayakkabı" , "Cosmetics" : "Makyaj malzemeleri" ,
                                    "Food & Beverage":"Yiyecek ve İçecek" , "Toys":"Oyuncak" , "Technology" : "Teknoloji" ,
                                    "Souvenir" : "Hatıra eşya"})
```

```python
df["odeme_turu"] = df["odeme_turu"].map({"Credit Card" : "Kredi Kartı" , "Debit Card" : "Banka Kartı" , "Cash" : "Nakit"})
```

```python
print("En Küçük Yaş :" , df["yas"].min())
print("En Büyük Yaş :" , df["yas"].max())
print("Ortalama Yaş :" , round(df["yas"].mean() , 2))

print("-------------------------------")

print("Ortalama Alınan Ürün Miktarı :" , round(df["miktar"].mean() , ))
```

```python
print("Kadınlara ve Erkeklere Ait En Düşük Yaş :" , df.groupby("cinsiyet").min()["yas"])
print("-------------")
print("Kadınlara ve Erkeklere Ait En Yüksek Yaş :" , df.groupby("cinsiyet").max()["yas"])
print("-------------")
print("Kadınlara ve Erkeklere Ait En Ortalama Yaş :" , round(df.groupby("cinsiyet").mean()["yas"] , 2))
```

```python
print("Kadınların ve Erkeklerin En Çok Gittiği AVM :" , df.groupby("cinsiyet").max()["avm"])
print("-----------------------")
print("Kadınların ve Erkeklerin En Az Gittiği AVM :" , df.groupby("cinsiyet").min()["avm"])
```

```python
print("Kadınların ve Erkeklerin En Çok Tercih Ettiği Kategori :" , df.groupby("cinsiyet").max()["kategori"])
print("-----------------------")
print("Kadınların ve Erkeklerin En Az Tercih Ettiği Kategori :" , df.groupby("cinsiyet").min()["kategori"])
```

```python
print("Kadınların ve Erkeklerin En Az Tercih Ettiği Ödeme Türü :" , df.groupby("cinsiyet").min()["odeme_turu"])
print("----------------------------")
print("Kadınların ve Erkeklerin En Fazla Tercih Ettiği Ödeme Türü :" , df.groupby("cinsiyet").max()["odeme_turu"])
```

```python
print("Kategorileri Göre Minimum Harcanan Para :" , round(df.groupby("kategori").min()["fiyat"] , 2))
print("------------------------")
print("Kategorileri Göre Maksimum Harcanan Para :" , round(df.groupby("kategori").max()["fiyat"] , 2))
print("------------------------")
print("Kategorileri Göre Ortalama Harcanan Para :" , round(df.groupby("kategori").mean()["fiyat"] , 2))
```

```python
print("AVM' lerde Minimum Harcanan Para : "  , df.groupby("avm").min()["fiyat"])
print("-----------------------")
print("AVM' lerde Maksimum Harcanan Para : "  , df.groupby("avm").max()["fiyat"])
print("------------------------")
print("AVM' lerde Ortalama Harcanan Para : "  , round(df.groupby("avm").mean()["fiyat"] , 2))
```

```python
print("Satın Alınan Kategorilere Ait Minimum Yaş :" , df.groupby("kategori").min()["yas"])
print("-------------")
print("Satın Alınan Kategorilere Ait Maksimum Yaş :" , df.groupby("kategori").max()["yas"])
print("-------------")
print("Satın Alınan Kategorilere Ait Ortalama Yaş :" , round(df.groupby("kategori").mean()["yas"] , 2))
```
