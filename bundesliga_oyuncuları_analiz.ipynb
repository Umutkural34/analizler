```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bundesliga_player.csv")
```

```python
df
```

```python
df.drop("full_name" , axis = 1 ,inplace = True)
```

```python
df
```

```python
df.columns = ["id","ad","yas","boy","ulke","dogum_yeri","piyasa_degeri","maks_piyasa_degeri","pozisyon","numara","ayak","kulup"
              ,"kontrat_tarihi","kulube_katilma","oyuncu_temsilcisi","malzemeci"]
```

```python
df.info()
```

```python
df["pozisyon"].unique()
```

```python
df["pozisyon"] = df["pozisyon"].map({"Goalkeeper" : "Kaleci" ,"Defender - Centre-Back" : "Stoper" , "Defender - Left-Back" : "Sol bek" ,
                                   "Defender - Right-Back" : "Sağ bek" , "midfield - Defensive Midfield" : "Defansif Orta Saha" ,
                                   "midfield - Central Midfiel" : "Merkez orta saha" , "midfield - Attacking Midfield" : "Ofansif orta saha" ,
                                   "Attack - Left Winger" : "Sol kanat" , "Attack - Right Winger" : "Sağ kanat" , "Attack - Second Striker" : "Sahte forvet" ,
                                   "Attack - Centre-Forward" : "Forvet" , "midfield - Right Midfield":"Sağ merkez" , "midfield - Left Midfield":"Sol merkez" })
```

```python
print("En Küçük Yaş : " , df["yas"].min())
print("En Büyük Yaş : " , df["yas"].max())
print("Ortalama Yaş : " , round(df["yas"].mean() , 2))
```

```python
print("En Küçük Boy : ", df["boy"].min())
print("En Büyük Boy : ", df["boy"].max())
print("Ortalama Boy : ", round(df["boy"].mean() , 2))
```

```python
print("En Düşük Piyasa Değeri : " , df["piyasa_degeri"].min() , "milyon €")
print("En Yüksek Piyasa Değeri : " , df["piyasa_degeri"].max() , " milyon €")
print("Ortalama Piyasa Değeri : " , round(df["piyasa_degeri"].mean() , 2) , "milyon €")
```

```python
print("Takımların Toplam Piyasa Değerleri : ")
for i in df["kulup"].unique():
    print(i , ":" , round(df[df["kulup"] == i]["piyasa_degeri"].sum() , 2) , "milyon €")
```

```python
print("Takımlarda Piyasa Değeri En Yüksek Oyuncular:")
for kulup, grp in df.groupby("kulup"):
    max_piyasa_degeri = grp["piyasa_degeri"].max()
    oyuncular = grp[grp["piyasa_degeri"] == max_piyasa_degeri]["ad"]
    for oyuncu in oyuncular:
        print("Takım:", kulup,"--------", "En Yüksek Piyasa Değerine Sahip Oyuncusu:", oyuncu)
```

```python
df
```

```python
print("Takımlara Ait En Düşük Boy :")
print(df.groupby("kulup").min()["boy"] )
print("----------------")
print("Takımlara Ait En Yüksek Boy :")
print(df.groupby("kulup").max()["boy"] )
print("----------------")
print("Takımlara Ait Boy Ortalaması :")
print(round(df.groupby("kulup").mean()["boy"] , 2))
```

```python
print("Pozisyonlara Göre En Düşük Piyasa Değeri : ")
print(df.groupby("pozisyon").min()["piyasa_degeri"])
print("----------------")
print("Pozisyonlara Göre En Yüksek Piyasa Değeri : ")
print(df.groupby("pozisyon").max()["piyasa_degeri"])
print("----------------")
print("Pozisyonlara Göre En Düşük Piyasa Değeri : ")
print(round(df.groupby("pozisyon").mean()["piyasa_degeri"],2))
```

```python
print("Pozisyonlara Göre En Küçük Piyasa Boy : ")
print(df.groupby("pozisyon").min()["boy"])
print("----------------")
print("Pozisyonlara Göre En Uzun Boy : ")
print(df.groupby("pozisyon").max()["boy"])
print("----------------")
print("Pozisyonlara Göre Ortalama Boy : ")
print(round(df.groupby("pozisyon").mean()["boy"],2))
```

```python
print("Poziyonlarında En Değerli Oyuncular:")
for pozisyon, grp in df.groupby("pozisyon"):
    max_piyasa_degeri = grp["piyasa_degeri"].max()
    oyuncular = grp[(grp["piyasa_degeri"] == max_piyasa_degeri)]["ad"]
    piyasa_degerleri = grp[(grp["piyasa_degeri"] == max_piyasa_degeri)]["piyasa_degeri"]
    for oyuncu, piyasa_degeri in zip(oyuncular, piyasa_degerleri):
        print("Pozisyon:", pozisyon, "- En Yüksek Piyasa Değeri:", piyasa_degeri, "- Oyuncu:", oyuncu)
```
